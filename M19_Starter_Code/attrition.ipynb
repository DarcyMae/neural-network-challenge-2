## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
attrition_df.columns
```

```python
from re import X
# Create a list of at least 10 column names to use as X data
columns_list = ['Education',
                'Age',
                'DistanceFromHome',
                'JobSatisfaction',
                'OverTime',
                'StockOptionLevel',
                'WorkLifeBalance',
                'YearsAtCompany',
                'YearsSinceLastPromotion',
                'NumCompaniesWorked'
       ]


# Create X_df using your selected columns
X_df = attrition_df[columns_list]


# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
X_train.head()
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train['OverTime'].value_counts()

from sklearn.preprocessing import OneHotEncoder
oh_encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
oh_encoder.fit(X_train[['OverTime']])
X_train['encoded_overtime']= oh_encoder.transform(X_train[['OverTime']])
X_test['encoded_overtime'] = oh_encoder.transform(X_test[['OverTime']])

X_train.drop(columns=['OverTime'], inplace=True)
X_test.drop(columns=['OverTime'], inplace=True)

print(X_train.dtypes)
print(X_test.dtypes)
```

```python
# Create a StandardScaler
StandardScaler = StandardScaler()

# Fit the StandardScaler to the training data
StandardScaler.fit(X_train)

# Scale the training and testing data
x_train_scaled = StandardScaler.transform(X_train)
x_test_scaled = StandardScaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
dept_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
dept_encoder.fit(np.array(y_train['Department']).reshape(-1, 1))


# Create two new variables by applying the encoder
# to the training and testing data
dept_train_encoder= dept_encoder.transform(np.array(y_train['Department']).reshape(-1, 1))
dept_test_encoder= dept_encoder.transform(np.array(y_test['Department']).reshape(-1, 1))

dept_train_encoder
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
attrition_encoder.fit(np.array(y_train['Attrition']).reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
attrition_train_encoder= attrition_encoder.transform(np.array(y_train['Attrition']).reshape(-1, 1))
attrition_test_encoder= attrition_encoder.transform(np.array(y_test['Attrition']).reshape(-1, 1))

attrition_train_encoder
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_dim = X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(input_dim,))

# Create at least two shared layers
shared_layer1 = layers.Dense(units=64, activation='relu', name='shared_layer1')(input_layer)
shared_layer2 = layers.Dense(units=128, activation='relu', name='shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden_layer = layers.Dense(units=32, activation='relu', name='dept_hidden_layer')(shared_layer2)

# Create the output layer
dept_output_layer = layers.Dense(units=3, activation='sigmoid', name='dept_output_layer')(dept_hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden_layer = layers.Dense(units=32, activation='relu', name='attrition_hidden_layer')(shared_layer2)

# Create the output layer
attrition_output_layer = layers.Dense(units=2, activation='sigmoid', name='attrition_output_layer')(attrition_hidden_layer)
```

```python
from ast import mod
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output_layer, attrition_output_layer])

# Compile the model
model.compile(optimizer='adam', loss={'dept_output_layer':'categorical_crossentropy',
                                      'attrition_output_layer':'binary_crossentropy'},
                                metrics={'dept_output_layer':'accuracy',
                                        'attrition_output_layer':'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(x_train_scaled, {'dept_output_layer': dept_train_encoder,
                           'attrition_output_layer':attrition_train_encoder}, epochs=100)
```

```python
# Evaluate the model with the testing data
results=model.evaluate(x_test_scaled, {'dept_output_layer': dept_test_encoder,
                               'attrition_output_layer':attrition_test_encoder})
results
```

```python
# Print the accuracy for both department and attrition
print('Department predictions accuracy: ', results[3])
print('Attrition predictions accuracy: ', results[4])
```
